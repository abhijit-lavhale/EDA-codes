```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]
plt.plot(x,y)
plt.title("Basic Line Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
plt.plot(x,y, color='red', linestyle ='--', marker = 'o',
         linewidth=1, markersize=4)
plt.title("Styled Line Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 10]

plt.bar(categories, values, color='orange')
plt.title("Bar chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
countries = ['India', 'USA', 'China', 'Germany', 'Brazil']
gdp = [3.7, 21.4, 17.8, 4.2, 2.1]
df = pd.DataFrame({'countries' : ['India', 'USA', 'China', 'Germany', 'Brazil'],
                   'gdp' : [3.7, 21.4, 17.8, 4.2, 2.1]})
df = df.sort_values('gdp', ascending = True)
df

plt.barh(df['countries'], df['gdp'], color = 'grey')
plt.title("GDP Bar Chart(horizontal)")
plt.xlabel("GDP")
plt.ylabel("Countries")
plt.show()
```

```python
# Random numerical data (e.g., age or income)
data = np.random.randn(1000)  # 1000 random numbers ~ Normal distribution
plt.hist(data, bins =20, color='orange', alpha=0.6, edgecolor= 'black')
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
plt.hist(df['Age'].dropna(), bins=20, color='teal',edgecolor='black')
plt.title("Age distribution of titanic passengers")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

Draw two overlapping histograms  one showing males ages and another showing females

```python
plt.hist(df[df['Sex'] == 'male']['Age'].dropna(), bins=20, alpha=0.5,edgecolor='black', label='Male')
plt.hist(df[df['Sex'] == 'female']['Age'].dropna(), bins=20, alpha=0.8,edgecolor='black', label='Female')
plt.title("Age Distribution by Gender")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
sns.histplot(data=df, x='Age', hue='Sex', bins=20, kde=True, alpha=0.5)
plt.title("Age Distribution by Gender (Seaborn)")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```
